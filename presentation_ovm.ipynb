```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def V(s):
    return 16.8 * (np.tanh(0.0860 * (s - 25)) + 0.913)

# Define the range of s (spacing)
s = np.linspace(0, 80, 400)
v = V(s)

# Compute thresholds
vmax = np.max(v)
low_speed_region = v < 3
high_speed_region = (vmax - v) < 3

# Plot
plt.figure(figsize=(7,4))
plt.plot(s, v, linewidth=2, color='black', label=r'$V(s)$')

# Fill entire vertical region based on x condition
ymin, ymax = 0, vmax + 5  # define fill range (slightly above curve for visibility)

# Red transparent region where V(s) < 3
plt.fill_between(s, ymin, ymax, where=low_speed_region, color='red', alpha=0.15,
                 label=r'$V(s) < 3$ m/s (Congested)')

# Green transparent region where Vmax - V(s) < 3
plt.fill_between(s, ymin, ymax, where=high_speed_region, color='green', alpha=0.15,
                 label=r'$V_{\max}-V(s) < 3$ m/s (Free flow)')

# Add reference lines
plt.axvline(x=25, color='red', linestyle='--', label='Critical spacing (25 m)')
plt.axhline(y=vmax, color='green', linestyle='--', label=fr'Max velocity ({vmax:.2f} m/s)')

# Labels and layout
plt.xlabel(r'Spacing $s$ (m)')
plt.ylabel(r'Optimal velocity $V(s)$ (m/s)')
plt.title('Optimal Velocity Function with Flow Regimes')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.ylim(ymin, ymax)
plt.tight_layout()

# Save and show
plt.savefig('optimal_velocity_regions.png', dpi=300)
plt.show()
```
